# 1. the top 10 countries with the highest total food production

```python
df['Total Food Production'] = df[['Maize Production (tonnes)', 'Rice  Production ( tonnes)', 'Wheat Production (tonnes)',
                                  'Yams  Production (tonnes)', 'Tomatoes Production (tonnes)', 'Sunflower seed  Production (tonnes)', 
                                  'Sugar cane Production (tonnes)', 'Soybeans  Production (tonnes)', 'Peas, dry Production ( tonnes)',
                                  'Palm oil  Production (tonnes)', 'Grapes  Production (tonnes)', 'Coffee, green Production ( tonnes)', 
                                  'Cocoa beans Production (tonnes)', 'Meat, chicken  Production (tonnes)', 'Bananas  Production ( tonnes)', 
                                  'Avocados Production (tonnes)', 'Apples Production (tonnes)', 'Sweet potatoes  Production (tonnes)',
                                  'Potatoes  Production (tonnes)']].sum(axis=1)
total_production_by_country = df.groupby('Entity')['Total Food Production'].sum()
top_10_countries = total_production_by_country.nlargest(10)

plt.figure(figsize=(10,6))
top_10_countries.plot(kind='bar', color='green')
plt.title('Top 10 Countries with Highest Total Food Production')
plt.xlabel('Country')
plt.ylabel('Total Food Production (tonnes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 2. the food production over the years

```python
df_grouped = df.groupby(['Year'])[['Total Food Production']].sum()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped, x='Year', y='Total Food Production', marker='o')
plt.title('Global Food Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.show()
```

# 3. the top 5 crops produced worldwide by production?

```python
total_productions = df[['Maize Production (tonnes)', 'Rice  Production ( tonnes)', 'Wheat Production (tonnes)',
                       'Yams  Production (tonnes)', 'Tomatoes Production (tonnes)', 'Sunflower seed  Production (tonnes)', 
                       'Sugar cane Production (tonnes)', 'Soybeans  Production (tonnes)', 'Peas, dry Production ( tonnes)',
                       'Palm oil  Production (tonnes)', 'Grapes  Production (tonnes)', 'Coffee, green Production ( tonnes)', 
                       'Cocoa beans Production (tonnes)', 'Meat, chicken  Production (tonnes)', 'Bananas  Production ( tonnes)', 
                       'Avocados Production (tonnes)', 'Apples Production (tonnes)', 'Sweet potatoes  Production (tonnes)',
                       'Potatoes  Production (tonnes)']].sum()
totals_df = pd.DataFrame(total_productions).reset_index()
totals_df.columns = ['Crop', 'Total Production']
total_production_dict = totals_df.set_index('Crop')['Total Production'].to_dict()
print("Total production dictionary:")
print(total_production_dict)
crops_data = {
    'Crop': total_productions.index,
    'Total Production': total_productions.values
}
crops_df = pd.DataFrame(crops_data)
totals_df = pd.DataFrame(total_productions).reset_index()
totals_df.columns = ['Crop', 'Total Production']
top_5_crops = totals_df.nlargest(5, 'Total Production')
plt.figure(figsize=(12, 8))
plt.barh(top_5_crops['Crop'], top_5_crops['Total Production'], color='skyblue')
plt.xlabel('Total Production (tonnes)')
plt.ylabel('Crop')
plt.title('Top 5 Crops by Total Production Worldwide')
plt.gca().invert_yaxis() 
plt.show()
```

# 4. Which countries show the highest increase in food production over the past decade?

```python
df['Total Food Production'] = df[['Maize Production (tonnes)', 'Rice  Production ( tonnes)', 'Wheat Production (tonnes)',
                                  'Yams  Production (tonnes)', 'Tomatoes Production (tonnes)', 'Sunflower seed  Production (tonnes)', 
                                  'Sugar cane Production (tonnes)', 'Soybeans  Production (tonnes)', 'Peas, dry Production ( tonnes)',
                                  'Palm oil  Production (tonnes)', 'Grapes  Production (tonnes)', 'Coffee, green Production ( tonnes)', 
                                  'Cocoa beans Production (tonnes)', 'Meat, chicken  Production (tonnes)', 'Bananas  Production ( tonnes)', 
                                  'Avocados Production (tonnes)', 'Apples Production (tonnes)', 'Sweet potatoes  Production (tonnes)',
                                  'Potatoes  Production (tonnes)']].sum(axis=1)
df_grouped = df.groupby(["Entity", "Year"])[["Total Food Production"]].sum().reset_index()
total_production_per_entity = df_grouped.groupby('Entity')['Total Food Production'].sum()
top_10_entities = total_production_per_entity.nlargest(10).index
df_top_10 = df_grouped[df_grouped['Entity'].isin(top_10_entities)]
df_pivot = df_top_10.pivot(index='Year', columns='Entity', values='Total Food Production')  

plt.figure(figsize=(14, 8))
for column in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[column], label=column)
plt.fill_between(df_pivot.index, df_pivot.sum(axis=1), alpha=0.3)  # Highlight the area
plt.title('Stacked Line Plot of Total Food Production by Top 10 Entities Across Years')
plt.xlabel('Year')
plt.ylabel('Total Food Production (tonnes)')
plt.legend(title='Entity')
plt.grid(True)
plt.show()
```

# 5. Disparities in Food Production Levels by Region

```python
df_region_grouped = df.groupby(["Entity", "Year"])[["Total Food Production"]].sum().reset_index()
df_disparities = df_region_grouped.groupby("Entity")["Total Food Production"].std().reset_index()
df_disparities.columns = ['Entity', 'Production Disparity']
df_disparities = df_disparities.sort_values('Production Disparity', ascending=False).head(10)
plt.figure(figsize=(12, 8))
sns.barplot(data=df_disparities, x='Production Disparity', y='Entity', palette='viridis')
plt.title('Top 10 Disparities in Food Production Levels by Region')
plt.xlabel('Standard Deviation of Total Food Production')
plt.ylabel('Region')
plt.grid(True)
plt.show()
```

# 6. Top 10 Crops in Top 10 Producing Countries by Total Production

```python
crop_columns = ['Maize Production (tonnes)', 'Rice  Production ( tonnes)', 'Wheat Production (tonnes)',
                 'Yams  Production (tonnes)', 'Tomatoes Production (tonnes)', 'Sunflower seed  Production (tonnes)', 
                 'Sugar cane Production (tonnes)', 'Soybeans  Production (tonnes)', 'Peas, dry Production ( tonnes)',
                 'Palm oil  Production (tonnes)', 'Grapes  Production (tonnes)', 'Coffee, green Production ( tonnes)', 
                 'Cocoa beans Production (tonnes)', 'Meat, chicken  Production (tonnes)', 'Bananas  Production ( tonnes)', 
                 'Avocados Production (tonnes)', 'Apples Production (tonnes)', 'Sweet potatoes  Production (tonnes)',
                 'Potatoes  Production (tonnes)']
df['Total Food Production'] = df[crop_columns].sum(axis=1)
top_producers = df.groupby('Entity')['Total Food Production'].sum().nlargest(10).index
top_producers_df = df[df['Entity'].isin(top_producers)]
melted_df = top_producers_df.melt(id_vars=['Entity'], value_vars=crop_columns, var_name='Crop', value_name='Production')
crop_totals = melted_df.groupby('Crop')['Production'].sum().nlargest(10).reset_index()
top_countries = melted_df[melted_df['Crop'].isin(crop_totals['Crop'])].groupby('Crop').apply(lambda x: x.loc[x['Production'].idxmax(), 'Entity']).reset_index()
top_countries.columns = ['Crop', 'Top Country']
crop_totals = crop_totals.merge(top_countries, on='Crop')

plt.figure(figsize=(14, 8))
barplot = sns.barplot(data=crop_totals, x='Production', y='Crop', palette='viridis')
for index, row in crop_totals.iterrows():
    barplot.text(row['Production'], index, f'{row["Production"]:.0f} ({row["Top Country"]})', 
                 color='black', ha='left', va='center')
plt.title('Top 10 Crops in Top 10 Producing Countries by Total Production')
plt.xlabel('Total Production (tonnes)')
plt.ylabel('Crop')
plt.grid(axis='x')
plt.show()
```
